## **About Dataset**

### **Dataset Overview:**

**The dataset provides a comprehensive set of features describing diamonds, which can be used to analyze, compare, and evaluate different diamonds based on their physical and grading characteristics. It includes aspects such as shape, color, clarity, and price, which are key factors in the diamond industry.**




### **Columns Explanation:**


1. Index: Unnamed index or identifier for each diamond entry.

2. Cut: The shape of the diamond (e.g., Round, Princess). Cut refers to how well the diamond is faceted, polished, and proportioned.

3. Color: Grading of the diamond’s color (e.g., E, F, G). The scale ranges from D (colorless) to Z (light yellow).

4. Clarity: Indicates the diamond's clarity, with grades like VVS2 (Very Very Slightly Included) to I (Included). It describes the inclusions or blemishes within the diamond.

5. Carat Weight: The weight of the diamond in carats. A key factor in determining a diamond’s size and value.

6. Cut Quality: Describes the quality of the diamond's cut, which affects its brilliance and sparkle (e.g., Excellent, Very Good).

7. Lab: The laboratory that graded the diamond (e.g., IGI, GIA). Different labs have varying standards.

8. Symmetry: Grade of the symmetry of the diamond’s facets (e.g., Very Good, Excellent). Well-aligned facets contribute to better light reflection.

9. Polish: The quality of the diamond's surface finish (e.g., Very Good, Excellent).

10. Eye Clean: Indicates whether inclusions are visible to the naked eye (e.g., Yes, No, Unknown).

11. Culet Size: Size of the diamond’s culet (point at the bottom of the diamond), ranging from None (no culet) to other descriptors.

12. Culet Condition: Condition of the culet, which can affect the diamond’s durability.

13. Depth Percent: Ratio of the diamond's depth to its diameter. It affects the diamond's appearance and light performance.

14. Table Percent: The width of the diamond’s top facet in relation to its overall width. It influences how light interacts with the diamond.

15. Measurements (Length, Width, Depth): Physical dimensions of the diamond.
Girdle Min/Max: The minimum and maximum thickness of the girdle, which is the edge of the diamond.

16. Fluorescence Color & Intensity: Fluorescence is the emission of light by the diamond when exposed to ultraviolet light. Can range from None to Very Strong.

17. Fancy Color Attributes: If the diamond has a fancy color (non-white color), these fields describe the dominant, secondary, and overtone colors as well as the color intensity.

18. Total Sales Price: The price at which the diamond was sold, in monetary units. This can be used for cost analysis and value assessment.


### **Possible Uses of the Dataset:**

*   Market Analysis: Understand trends in diamond quality, pricing, and preferred attributes.
*   Pricing Models: Build models to predict diamond prices based on various attributes.
*   Quality Control: Identify factors that affect the overall quality and value of diamonds.
*   Consumer Preferences: Analyze which features (like color, clarity, carat) are more desirable among buyers.

```python
import pandas as pd

# Define a function to remove outliers using IQR method
def remove_outliers_iqr(df):
    # Loop through each column to process numerical columns only
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Define outlier boundaries
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out the outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

    return df

# Apply the function to remove outliers
cleaned_data = remove_outliers_iqr(df)

# Show the shape before and after removing outliers
original_shape = df.shape
cleaned_shape = cleaned_data.shape

original_shape, cleaned_shape
```

# Dimentionality Reduction

## PCA Method

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the features before PCA
features = ['color', 'clarity', 'symmetry', 'carat_weight']

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(df[features])
#print(X)

# Apply PCA
pca = PCA()
pca.fit(X)

# Explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(features) + 1), pca.explained_variance_ratio_, marker='o')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid()
plt.show()

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], hue=df['total_sales_price'], palette='viridis')
plt.title('PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Total Sales Price')
plt.show()
```

# Conclusion for Diamond Price Prediction Project

This project aimed to predict diamond prices using various features such as carat, cut, color, clarity, and others. The data preprocessing phase was comprehensive, involving several important steps to optimize the dataset for modeling. Categorical features were encoded into numerical values using Label Encoding, and a correlation analysis was conducted to drop features with low correlation to the target variable. To further enhance model performance, Principal Component Analysis (PCA) was applied to reduce dimensionality and retain the most important features. Data normalization techniques such as StandardScaler and MinMaxScaler were employed to ensure that all features were on a comparable scale, preventing bias toward features with larger numerical ranges.

With the data preprocessed and optimized, various regression models were applied to predict diamond prices:

* Linear Regression
* Ridge Regression
* Lasso Regression
* Polynomial Regression
* Decision Tree Regression
* Random Forest Regression
* Gradient Boosting

The Decision Tree Regression model stood out as the most effective, delivering the lowest error and the highest R-squared score. It provided the best balance of interpretability and predictive power, making it the most suitable model for this project.

Through a combination of data preprocessing, dimensionality reduction, outlier removal, scaling, and advanced regression techniques, this project successfully built a robust diamond price prediction model. The insights gained provide valuable tools for stakeholders in the diamond industry to understand pricing dynamics and make data-driven decisions.
